```python
import json
import pandas as pd

with open("marques_brownlee.json","r") as f:
    file_data = json.load(f)

channel_id, statistics = file_data.popitem()
print(channel_id)
channel_stats = statistics["channel_statistics"]
video_stats = statistics["video_data"]
```

```python
#Channel statistics
view_count = channel_stats["viewCount"]
subscriber_count = channel_stats["subscriberCount"]
video_count = channel_stats["videoCount"]
print(f"View count: {view_count}")
print(f"Subscriber_count: {subscriber_count}")
print(f"Video_count: {video_count}")
```

```python
# Video Statistics
sorted_videos = sorted(video_stats.items(), key =lambda item : int(item[1]["viewCount"]), reverse=True)
print(f"Sorted videos length: {len(sorted_videos)}")
```

```python
video_statistics = list()
for video in sorted_videos:
    video_id = video[0]
    title = video[1]["title"]
    view_count = int(video[1]["viewCount"])
    like_count = int(video[1]["likeCount"])
    fav_count = int(video[1]["favoriteCount"])
    comment_count = int(video[1]["commentCount"])
    dislike_count = int(video[1]["dislikeCount"])
    video_statistics.append([title,view_count,like_count,fav_count,comment_count,dislike_count])
    
df = pd.DataFrame(video_statistics, columns=["Title","Views","Likes","Comments","Dislikes","Favourites"])

```

```python
# Top ten vidoes with most views
top_ten = df.head(10)
print(top_ten)
graph = top_ten.plot.bar(x="Title",y="Views",figsize=(13,9), fontsize=14)
```

```python
# Bottom ten vidoes with most views
bottom_ten = df.tail(10)
print(bottom_ten)
graph = bottom_ten.plot.bar(x="Title",y="Views",figsize=(13,9), fontsize=14)
```

```python
#Likes/Dislikes average of Top ten
likes_avg = top_ten["Likes"].mean()
print(f"TOp 10 videos likes average : {likes_avg}")
dislikes_avg = top_ten["Dislikes"].mean()
print(f"TOp 10 videos dislikes average: {dislikes_avg}")
```

```python
#Likes/Dislikes average of all videos
all_likes_avg = df["Likes"].mean()
print(f"All Videos likes average : {all_likes_avg}")
all_dislikes_avg = df["Dislikes"].mean()
print(f"All videos dislikes average : {all_dislikes_avg}")
```
